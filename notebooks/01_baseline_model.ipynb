# Baseline model: 60-min forward returns

This notebook builds a simple baseline on 5-minute bars. It assumes you ingested TRADES ticks.

```python
import polars as pl
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score

ticks = pl.scan_parquet('data/raw/ticks/source=ib/**/*.parquet')
ticks = ticks.select(['ts', 'symbol', 'last', 'size'])
ticks.head(5).collect()
```

```python
# Resample to 5-minute bars
bars = (
    ticks
    .group_by_dynamic('ts', every='5m', by='symbol', closed='left')
    .agg([
        pl.col('last').last().alias('price'),
        pl.col('last').mean().alias('price_mean'),
        pl.col('size').sum().alias('vol'),
        pl.col('last').std().alias('price_std'),
    ])
)
bars = bars.sort(['symbol', 'ts'])
bars.head(5)
```

```python
# Feature engineering on 5-min bars
bars = bars.with_columns([
    (pl.col('price') / pl.col('price').shift(1) - 1).over('symbol').alias('ret_5m'),
    pl.col('price').pct_change().over('symbol').alias('ret_5m_alt'),
    pl.col('vol').rolling_mean(12).over('symbol').alias('vol_1h'),
    pl.col('price_std').rolling_mean(12).over('symbol').alias('std_1h'),
])
# Forward return: 60 minutes ahead = 12 * 5-minute bars
bars = bars.with_columns((pl.col('price').shift(-12) / pl.col('price') - 1).over('symbol').alias('ret_fwd_60m'))
bars = bars.drop_nulls()
bars.select(['symbol','ts','ret_5m','vol_1h','std_1h','ret_fwd_60m']).head(5)
```

```python
# Train/test split by time (last 20% for test)
df = bars.collect().to_pandas()
df = df.sort_values(['symbol','ts'])
cut = int(len(df) * 0.8)
train = df.iloc[:cut]
test = df.iloc[cut:]

features = ['ret_5m', 'vol_1h', 'std_1h']
X_train = train[features].fillna(0.0).values
y_train = train['ret_fwd_60m'].values
X_test = test[features].fillna(0.0).values
y_test = test['ret_fwd_60m'].values

model = Ridge(alpha=1.0)
model.fit(X_train, y_train)
pred = model.predict(X_test)
print('R2:', r2_score(y_test, pred))
print('Correlation:', np.corrcoef(y_test, pred)[0,1])
```

## Optional: PyTorch baseline
A tiny linear model trained with MSE.

```python
import torch
from torch import nn

X_train_t = torch.tensor(X_train, dtype=torch.float32)
y_train_t = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
X_test_t = torch.tensor(X_test, dtype=torch.float32)

model = nn.Linear(X_train.shape[1], 1)
opt = torch.optim.Adam(model.parameters(), lr=1e-2)
loss_fn = nn.MSELoss()

for epoch in range(200):
    opt.zero_grad()
    out = model(X_train_t)
    loss = loss_fn(out, y_train_t)
    loss.backward()
    opt.step()

with torch.no_grad():
    pred_t = model(X_test_t).numpy().reshape(-1)

print('Torch R2:', r2_score(y_test, pred_t))
print('Torch Corr:', np.corrcoef(y_test, pred_t)[0,1])
```
